```python
# importing all the required libraries
from bs4 import BeautifulSoup
import requests,openpyxl
```

```python
# for creating an excel file
excel=openpyxl.Workbook()
sheet=excel.active
sheet.title='IMDB Top rated Movies'
print(excel.sheetnames)
sheet.append(['Movie Rank','Movie Name','Year of Release','Rating'])
```

```python
#accessing to a particlar website's information
# we will be extracting data from IMDB Website
source= requests.get('https://www.imdb.com/chart/top/')
source.raise_for_status() # checking error in our url
```

```python
soup=BeautifulSoup(source.text,'html.parser') # to collect html source code
# print(soup)
```

```python
movies=soup.find('tbody',class_='lister-list').find_all('tr') # to access all the list
print(len(movies))
```

```python
# to extract all the movie names, ranks,years,and ratings
for movie in movies:
    name=movie.find('td',class_='titleColumn').a.text
    rank=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
    year=movie.find('td',class_='titleColumn').span.text.strip('()')
    rating=movie.find('td',class_='ratingColumn imdbRating').strong.text
    sheet.append([rank,name,year,rating]) # append this list with excel
    print(name,rank,year,rating)
    
    
    
```

```python
excel.save('IMDB Movie Ratings.xlsx') # save everything as an excel file
```
